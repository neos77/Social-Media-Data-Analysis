```python
# 1 IMPORTING LIBRARIES.

import seaborn as sns  
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import random 
```

```python
# 2 CREATING DATA FOR THE PROJECT.

catagory=["Food", "Travel", "Fashion", "Fitness", "Music", "Culture", "Family","Health" ]

data={ 'Date':pd.date_range('2022-01-01',periods=700),
       'Catagory':[random.choice(catagory) for _ in range(700)],
       'Likes':np.random.randint(0,100000,size=700)
    }
```

```python
# 3 EXPLORING THE DATA
df=pd.DataFrame(data)
print(df)
print("\n1st 7 data:")
print(df.head(7))
print("\nlast 7 data:")
print(df.tail(7))
print("\ninformation:")
df.info()
print("\ndescribtion:")
print(df.describe())
print("\ncatagories count:")
df.value_counts("Catagory")
```

```python
# 4 CLEANING

check=df.isnull().any().any()
if check == True:
    print("\n null values exists in the data")
else:
    print("\n no null values in the data")
    
df.drop_duplicates(keep=False)                     # removing all duplicated values.
pd.to_datetime(df['Date'])                         # Converting the 'Date' column to datetime format.
df.astype({'Likes':'int32'}).dtypes                #converting the dataframe ‘Likes’ data to an integer
```

```python
# 5  VISUALIZATION AND ANALYSATION
print("Visulization of the Data")
print("\n")
print(sns.histplot(df['Likes']))
```

```python
sns.boxplot(data=data, x="Likes", y="Catagory")
```
